# Calculus Implementation

```python
B = gambar[:,:,2]
B_dx,B_dy = np.gradient(B)
fig, ax = plt.subplots(ncols=3,figsize=(21,3))
ax[0].imshow(B)
ax[0].set_title('Gambar Filter B')
ax[1].imshow(B_dx)
ax[1].set_title('dx')
ax[2].imshow(B_dy)
ax[2].set_title('dy')
```

```python
B_dx
```

```python
B_dy
```

# Menghitung gradient magnitude tiap pixel

```python
magnitude = np.sqrt ((B_dx**2) + (B_dy**2))
plt.figure(figsize = (21,5))
plt.imshow(magnitude, cmap='gray')
print('max', magnitude.max())
print('min', magnitude.min())
```

# Menghitung sudut tiap pixel

```python
Angle = np.rad2deg(np.arctan2(B_dy, B_dx)) + 180
plt.figure(figsize = (21,5))
plt.imshow(Angle, cmap='gray')
print('max', Angle.max())
print('min', Angle.min())
```

# Threshold

Konversi numpy array ke dalam threshold

```python
thres_30 = np.where(magnitude > 30, 0, 30)
```

array yang besar dari 30 akan dibuat 0 dan kecil dari 30 akan dibuat 30, jadi nilai array yang paling besar adalah 30

```python
thres_30
```

```python
plt.figure(figsize = (21,5))
plt.imshow(thres_30, cmap='gray')
print('max', thres_30.max())
print('min', thres_30.min())
```

```python
thres_60 = np.where(magnitude > 60, 0, 60)
```

```python
thres_60
```

```python
plt.figure(figsize = (21,5))
plt.imshow(thres_60, cmap='gray')
print('max', thres_60.max())
print('min', thres_60.min())
```

```python
thres_90 = np.where(magnitude > 90, 0, 90)
```

```python
fig, ax = plt.subplots(ncols=4,figsize=(21,3))
ax[0].imshow(thres_30, cmap='gray')
ax[0].set_title('Threshold 30')
ax[1].imshow(thres_60, cmap='gray')
ax[1].set_title('Threshold 60')
ax[2].imshow(thres_90, cmap='gray')
ax[2].set_title('Threshold 90')
ax[3].imshow(hasil_perkalian_matriks, cmap='gray')
ax[3].set_title('Hasil perkalian matriks')
```

# Overall Analysis

Perbandingan antar 3 gambar diatas yang memiliki threshold berbeda : 
- untuk threshold dengan nilai max 30, deteksi tepi yang dihasilkan untuk objek cukup jelas walaupun ada beberapa garis tepi yang hilang dibagian kepala dikarenakan gambar asli dibagian tersebut gelap
- untuk threshold dengan nilai max 60, deteksi tepi yang dihasilkan untuk objek kurang jelas banyak garis tepi yang hilang dibagian objek dikarenakan threshold dinaikkan dan gambar semakin terang
